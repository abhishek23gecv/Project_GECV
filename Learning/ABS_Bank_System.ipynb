```

class Bank:
    def __init__(self):
        self.accounts = {}

    def create_account(self, account_number, account_holder):
        if account_number in self.accounts:
            print(f"\nâš ï¸ Account number {account_number} already exists.")
        else:
            self.accounts[account_number] = {
                'holder': account_holder,
                'balance': 0
            }
            print(f"\nâœ… Account created successfully for {account_holder} (Account No: {account_number}).")

    def deposit(self, account_number, amount):
        if account_number in self.accounts:
            self.accounts[account_number]['balance'] += amount
            print(f"\nğŸ’° Deposited â‚¹{amount} to account {account_number}. New balance: â‚¹{self.accounts[account_number]['balance']}")
        else:
            print(f"\nâŒ Account number {account_number} does not exist.")

    def withdraw(self, account_number, amount):
        if account_number in self.accounts:
            if self.accounts[account_number]['balance'] >= amount:
                self.accounts[account_number]['balance'] -= amount
                print(f"\nğŸ’¸ Withdrew â‚¹{amount} from account {account_number}. Remaining balance: â‚¹{self.accounts[account_number]['balance']}")
            else:
                print(f"\nâš ï¸ Insufficient balance in account {account_number}.")
        else:
            print(f"\nâŒ Account number {account_number} does not exist.")

    def check_balance(self, account_number):
        if account_number in self.accounts:
            print(f"\nğŸ“Š Balance for account {account_number}: â‚¹{self.accounts[account_number]['balance']}")
        else:
            print(f"\nâŒ Account number {account_number} does not exist.")

    def get_account_info(self, account_number):
        if account_number in self.accounts:
            account = self.accounts[account_number]
            print("\nğŸ§¾ Account Information:")
            print(f"ğŸ”¢ Account Number: {account_number}")
            print(f"ğŸ‘¤ Account Holder: {account['holder']}")
            print(f"ğŸ’¼ Balance: â‚¹{account['balance']}")
        else:
            print(f"\nâŒ Account number {account_number} does not exist.")


# ---------- Helper Functions ----------

def display_landing():
    print("=" * 50)
    print("ğŸ¦ Welcome to ABS CLI - Your Simple Bank System")
    print("=" * 50)

def display_menu():
    print("\nğŸ“‹ Menu:")
    print("1ï¸âƒ£  Create Account")
    print("2ï¸âƒ£  Deposit Money")
    print("3ï¸âƒ£  Withdraw Money")
    print("4ï¸âƒ£  Check Balance")
    print("5ï¸âƒ£  View Account Info")
    print("6ï¸âƒ£  Exit")

def get_account_number():
    return input("ğŸ”¢ Enter account number: ").strip()

def get_amount_input(prompt):
    while True:
        try:
            amount = float(input(prompt))
            if amount < 0:
                print("âš ï¸ Amount cannot be negative. Try again.")
            else:
                return amount
        except ValueError:
            print("âš ï¸ Invalid input. Please enter a number.")

# ---------- Main Program ----------

bank = Bank()
display_landing()

while True:
    display_menu()
    choice = input("\nğŸ“ Enter your choice (1-6): ").strip()

    if choice == '1':
        acc_num = get_account_number()
        acc_holder = input("ğŸ‘¤ Enter account holder name: ").strip()
        bank.create_account(acc_num, acc_holder)

    elif choice == '2':
        acc_num = get_account_number()
        amount = get_amount_input("ğŸ’° Enter amount to deposit: â‚¹")
        bank.deposit(acc_num, amount)

    elif choice == '3':
        acc_num = get_account_number()
        amount = get_amount_input("ğŸ’¸ Enter amount to withdraw: â‚¹")
        bank.withdraw(acc_num, amount)

    elif choice == '4':
        acc_num = get_account_number()
        bank.check_balance(acc_num)

    elif choice == '5':
        acc_num = get_account_number()
        bank.get_account_info(acc_num)

    elif choice == '6':
        print("\nğŸ‘‹ Thank you for using ABS. Have a great day!")
        break

    else:
        print("\nâ— Invalid choice. Please enter a number between 1 and 6.")
```
